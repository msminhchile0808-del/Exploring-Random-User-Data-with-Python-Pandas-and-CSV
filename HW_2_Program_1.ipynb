```python
import requests
import json

# 1. Read full info for 25 people from the randomuser api

api_url = requests.get("https://randomuser.me/api/?results=25")
print(api_url)
```

```python
result = json.loads(api_url.content)
print(result)
```

```python
print(json.dumps(result, indent=2))
```

```python
import pandas as pd

# 2. Select the gender, title, first, last names, email, location, nationality, and picture URLs
# 3. Convert results into a pandas dataframe

my_df = pd.DataFrame(result['results'])
my_df = my_df[["gender", "name", "email", "location", "nat", "picture"]]
my_df
```

```python
# 4. Store the results into a CSV file
my_df.to_csv('random_users.csv', index=False)
print("Data saved to random_users.csv")
```

```python
# 5. Using crosstabs, show the counts of people by gender by nationality

df_cross = pd.crosstab(my_df['gender'],my_df['nat'])
print(df_cross)
```
