```python
import requests
import json

# 1. Read full info for 25 people from the randomuser api

api_url = requests.get("https://randomuser.me/api/?results=25&gender=female&nat=US&inc=gender,name,email,location,nat,picture")
print(api_url)
```

```python
result = json.loads(api_url.content)
print(result)
```

```python
print(json.dumps(result, indent=2))
```

```python
import pandas as pd

# Convert the 'results' list from the dictionary into a pandas DataFrame
df_users = pd.DataFrame(result['results'])
print(df_users.head(5))
```

```python
# Save the DataFrame to a CSV file
df_users.to_csv('random_users.csv', index=False)
print("Data saved to random_users.csv")
```
